```python
rel = "hf://datasets/FlyRank/internship-warehouse"

con.sql(f"""
SELECT
    COUNT(*) AS rows,
    MIN(report_date) AS start_date,
    MAX(report_date) AS end_date,
    COUNT(DISTINCT report_date) AS days,
    COUNT(DISTINCT client_hash_id) AS clients,
    COUNT(DISTINCT content_hash_id) AS contents
FROM read_parquet('{rel}/fact_content_daily_performance/**/*.parquet')
WHERE month = '2026-03'
""").df()
```

```python
rel = "hf://datasets/FlyRank/internship-warehouse"

con.sql(f"""
SELECT
    COUNT(*) AS total_rows,
    COUNT(*) FILTER (WHERE gsc_data_available IS TRUE) AS gsc_available,
    COUNT(*) FILTER (WHERE ga4_data_available IS TRUE) AS ga4_available
FROM read_parquet('{rel}/fact_content_daily_performance/**/*.parquet')
WHERE month = '2026-03'
""").df()
```

## 2. Fields: feature / label / context / excluded

## Fields

### Features
- gsc_impressions
- gsc_clicks
- gsc_sum_position
- client_has_gsc
- gsc_data_available

These fields are used as input features because they describe the observed search performance of a page.

### Label
The label (proxy) is whether a page is declining. The decline label represents the outcome the model would predict for this lane..

### Context
- report_date
- client_hash_id
- content_hash_id
- month

These fields identify the page, client and time period.

### Excluded
GA4 metrics are excluded because many rows do not have GA4 data available, which would reduce the usable dataset.

```python
rel = "hf://datasets/FlyRank/internship-warehouse"

con.sql(f"""
SELECT
COUNT(*) AS total_rows,
COUNT(*) FILTER (WHERE gsc_data_available IS TRUE) AS gsc_available,
COUNT(*) FILTER (WHERE ga4_data_available IS TRUE) AS ga4_available
FROM read_parquet('{rel}/fact_content_daily_performance/**/*.parquet')
WHERE month='2026-03'
""").df()
```

## 3. Verify it with queries (grain, counts, missing values, windows)

## Verification

The following queries verify the contract:

## 4. Data limits

## Data Limits

This dataset is anonymized and does not contain client identities or page content.

The historical coverage differs between clients, so it is an unbalanced panel.

Many rows do not contain GA4 data, therefore analyses relying on GA4 metrics may use fewer observations.

The dataset supports decision-making but cannot explain every reason behind changes in page performance.

```python
rel = "hf://datasets/FlyRank/internship-warehouse"

con.sql(f"""
SELECT
COUNT(*) FILTER (WHERE ga4_data_available IS TRUE) AS ga4_available
FROM read_parquet('{rel}/fact_content_daily_performance/**/*.parquet')
WHERE month='2026-03'
""").df()
```
